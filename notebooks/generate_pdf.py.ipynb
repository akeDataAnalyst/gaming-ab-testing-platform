```python
from fpdf import FPDF

pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)

title = "Executive Summary – Gaming A/B Test"
pdf.set_font("Arial", 'B', 16)
pdf.cell(200, 10, txt=title, ln=1, align='C')

pdf.set_font("Arial", size=12)
pdf.ln(10)

content = """
Variant B (streak rewards) shows strong, statistically significant uplifts:

• D7 Retention: +30.5% (p << 0.001)
• D30 Retention: +58.1% (p << 0.001)
• Revenue per user: +18.4% (p << 0.001)
• Payer conversion: +68.3% (p << 0.001)

Recommendation: Roll out Variant B to all users.

Prepared by Aklilu Abera
"""

pdf.multi_cell(0, 10, content)

pdf.output("executive_summary.pdf")
print("PDF created: executive_summary.pdf")
```
